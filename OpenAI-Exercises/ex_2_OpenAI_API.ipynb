**Traccia:** Crea un prompt che estragga i dettagli chiave (nome, età, professione) da un testo descrittivo su una persona.

**Obiettivo:** Estrarre informazioni strutturate da un testo libero.

```python

# TODO Inizializza il client di Open AI
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env') # Read the .env file

# Configure OpenAI client
openai.api_key = os.environ['OPENAI_API_KEY']

client = OpenAI()

```

```python
input_text = input("Enter text: ")

complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{
            "role": "system",
            "content": (
                "You are an information extraction assistant. "
                "Extract the following details from the text: "
                "name, age, profession. "
                "Return the result in a structured format."
            )
        },
        {
            "role": "user",
            "content": input_text
        }],
)

print(complention.choices[0].message.content)
```
