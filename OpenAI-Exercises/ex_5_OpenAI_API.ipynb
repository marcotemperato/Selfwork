```python

# TODO Inizializza il client di Open AI
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')

openai.api_key = os.environ["OPENAI_API_KEY"]

client = openai.OpenAI()
```

```python
input_text = input("Enter text: ")


complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "system", "content": "Fornito un testo di input, comprendilo e restituisci delle domande per testare la comprensione del testo stesso."},
              {"role": "user", "content": f"Genera 3 domande per testare la comprensione "
                f"del seguente testo:\n\n{input_text}"
              },
             ]
)

print(complention.choices[0].message.content)
```
