```python

# TODO Inizializza il client di Open AI
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')

openai.api_key = os.environ['OPENAI_API_KEY']

client = openai.OpenAI()

```

```python
input_text = input("Enter text: ")
data = {"name": "Giovanni", "age": 30, "city": "Roma", "profession": "Ingegnere"};


complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": input_text},
              {"role": "system", "content": f"Data: {data}"}],
)

print(complention.choices[0].message.content)
```
