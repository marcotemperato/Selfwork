```python

# TODO Inizializza il client di Open AI
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env') # Read the .env file

# Configure OpenAI client
openai.api_key = os.environ['OPENAI_API_KEY']

client = OpenAI()
```

```python
input_text = input("Enter text: ")

complention = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "system", "content": "Riassumi il seguente testo in massimo 255 caratteri."},
              {"role": "user", "content": input_text}],
)

print(complention.choices[0].message.content)
```
