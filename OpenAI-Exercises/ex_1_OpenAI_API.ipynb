```python

import os
from dotenv import load_dotenv
from openai import OpenAI

# Carica file .env
load_dotenv()

# Recupera chiave API
api_key = os.getenv("OPENAI_API_KEY")

print(api_key)

# Crea client
client = OpenAI(api_key=api_key)
```

```python
input_text = input("Enter text to translate (Italian to English or English to Italian): ")

completion = client.chat.completions.create(
    model="gpt-5",
    messages=[
        {
            "role": "system",
            "content": "You are a professional translator. Translate texts from Italian to English and from English to Italian."
        },
        {
            "role": "user",
            "content": input_text
        }
    ],
)

print(completion.choices[0].message.content)
```
