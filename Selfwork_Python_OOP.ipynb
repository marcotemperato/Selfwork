```python
#Esercizi:
#Creare una classe Person, estenderla con due sottoclassi Dipendente e Studente
#Estendere Dipendenti con 2 sottoclassi Docente e Freelance
#Una classe deve essere astratta e ogni classe deve avere delle proprietà specializzanti

from abc import ABC, abstractmethod
class Person(ABC):
    def __init__(self, name, age):
        self.name = name
        self.age = age

    @abstractmethod
    def get_info(self):
        pass
    
class Dipendente(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def get_info(self):
        return f"Dipendente: {self.name}, Age: {self.age}, Salary: {self.salary}"
    
class Studente(Person):
    def __init__(self, name, age, course):
        super().__init__(name, age)
        self.course = course

    def get_info(self):
        return f"Studente: {self.name}, Age: {self.age}, Course: {self.course}"
    
class Docente(Dipendente):
    def __init__(self, name, age, salary, subject):
        super().__init__(name, age, salary)
        self.subject = subject

    def get_info(self):
        return f"Docente: {self.name}, Age: {self.age}, Salary: {self.salary}, Subject: {self.subject}"
    
class Freelance(Dipendente):
    def __init__(self, name, age, salary, project):
        super().__init__(name, age, salary)
        self.project = project

    def get_info(self):
        return f"Freelance: {self.name}, Age: {self.age}, Salary: {self.salary}, Project: {self.project}"
    
#Esempio di utilizzo
docente = Docente("Mario Rossi", 45, 3000, "Matematica")
freelance = Freelance("Luigi Bianchi", 30, 2000, "Sviluppo Web")
studente = Studente("Anna Verdi", 20, "Informatica")
print(docente.get_info())
print(freelance.get_info())
print(studente.get_info())
```
