**Traccia:** Trasforma un set di dati strutturati (come un dizionario Python) in una descrizione testuale completa.

**Obiettivo:** Convertire dati strutturati in una descrizione leggibile.

```python
import ollama

data = {"nome": "Marco", "età": 34, "professione": "Sviluppatore"}

response = ollama.chat(
    model="llama3",
    messages=[{"role": "system", "content": "Trasforma i dati del dizionario in un testo descrittivo, leggibile e in lingua italiana."},
              {"role": "user", "content": "Trasforma questi dati in un testo: " + str(data)}]
)

print(response['message']['content'])
```
