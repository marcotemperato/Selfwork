**Traccia:** Scrivi un prompt che generi domande di comprensione su un breve testo.

**Obiettivo:** Sviluppare un prompt per creare domande educative basate su un contenuto.

```python
import ollama

input_text = input("Inserisci un testo:")

response = ollama.chat(
    model="llama3",
    messages=[{"role": "system", "content": "Fornito un testo di input, comprendilo e restituisci delle domande per testare la comprensione del testo stesso"},
              {"role": "user", "content": input_text}]
    
)

print(response['message']['content'])
```
